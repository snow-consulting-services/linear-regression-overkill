# Why not use a neural network for everything? :-) 

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
n = 100

x = np.hstack([np.ones(n).reshape(n,1), np.linspace(0,1,n).reshape(n,1)])
y = np.array([i + np.random.rand() for i in x[:,1]])

w = np.array([0, 0])

# iterations
k = 10000
epochs = list(range(0,k))

# learning rate
eps = 1e-3

for epoch in epochs:
    grad = np.array([0., 0.])
    
    for i in range(0, n):
        x_i = x[i, :]
        y_i = y[i]
        h = np.dot(w, x_i) - y_i
        grad += 2*x_i*h
    w = w - eps*grad

plt.scatter(x[:, 1], y)
plt.plot(x[:,1], w[1]*x[:,1] + w[0])

plt.show()
```

```python
print('intercept: ' + str(w[0]))
print('gradient: ' + str(w[1]))
```

# Exact solution 

```python
lr = LinearRegression().fit(x[:,1].reshape(-1,1), y)
print('intercept: ' + str(lr.intercept_))
print('gradient: ' + str(lr.coef_[0]))
```

```python
plt.scatter(x[:,1], y)
plt.plot(x[:,1], lr.coef_[0] * x[:,1] + lr.intercept_)
```
